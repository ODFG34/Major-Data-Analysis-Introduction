## Naive Bayesian Classification

### Bayes' theorem

$$ P(A\mid B)={\frac {P(B\mid A)\,P(A)}{P(B)}} $$

Using the Bayes' theorem, probability of label given some observed features: $$ P(L~|~{\rm features}) = \frac{P({\rm features}~|~L)P(L)}{P({\rm features})} $$ 

If we are trying to decide between two labels: $L_1$, $L_2$, then one way to make this decision is to compute the ratio of the posterior probabilities for each label:

$$ \frac{P(L_1~|~{\rm features})}{P(L_2~|~{\rm features})} = \frac{P({\rm features}~|~L_1)}{P({\rm features}~|~L_2)}\frac{P(L_1)}{P(L_2)} $$

# Gaussian Naive Bayes

Let's generate synthetic data

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu');
```

Let's use scikit-learn 

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, y);
```

Let's test our model! 

```python
rng = np.random.RandomState(0)
Xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2)
ynew = model.predict(Xnew)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
lim = plt.axis()
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20, cmap='RdBu', alpha=0.1)
plt.axis(lim);
```

Can we guess the probability of predictions ? 

```python
yprob = model.predict_proba(Xnew)
```

```python
yprob
```

```python
yprob.round(2)
```

```python
messages['label_num'] = messages.label.map({'ham':0, 'spam':1})
```

```python
messages.head()
```

## Feature Engineering 

```python
## Counting how many times a word appears in the dataset
import pandas as pd
from collections import Counter
text = pd.DataFrame(messages['message'])

total_counts = Counter()
for i in range(len(text)):
    for word in text.values[i][0].split(" "):     ## split sentece into words
        total_counts[word] += 1

print("Total words in data set: ", len(total_counts))
```

```python
total_counts
```

```python
## Sorting in decreasing order (Word with highest frequency appears first)
vocab = sorted(total_counts, key=total_counts.get, reverse=True)
print(vocab[:60])
```

```python
# Mapping from words to index

vocab_size = len(vocab)
word2idx = {}
#print vocab_size
for i, word in enumerate(vocab):
    word2idx[word] = i
```

```python
### Text to Vector
def text_to_vector(text):
    word_vector = np.zeros(vocab_size)
    for word in text.split(" "):
        if word2idx.get(word) is None:
            continue
        else:
            word_vector[word2idx.get(word)] += 1
    return np.array(word_vector)
```

```python
## Convert all titles to vectors
import numpy as np
word_vectors = np.zeros((len(text), len(vocab)), dtype=np.int_)
for ii, (_, text_) in enumerate(text.iterrows()):
    word_vectors[ii] = text_to_vector(text_[0])
```
