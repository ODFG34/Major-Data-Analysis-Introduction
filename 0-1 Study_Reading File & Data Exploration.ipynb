# Dataframe Plotting

To use pyplot, follow the commands below: 

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

Now, let's plot it. 

```python
plot_data = df[df['Element'] == 'Food']
plot_data = plot_data.groupby('Area')['Y2013'].sum()
plot_data.sort_values()[:10].plot(kind='bar')   
```
